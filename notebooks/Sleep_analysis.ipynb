```python
#preparing correlation df for a heatmap
sleep_analysis.columns
sleep_analysis.info()
```

```python
df_corr = sleep_analysis[['Age', 'Sleep Duration',
       'Quality of Sleep', 'Physical Activity Level', 'Stress Level', 'Heart Rate', 'Daily Steps']].corr()
df_corr
```

```python
fig, ax = plt.subplots(figsize=(10,10))
sns.heatmap(df_corr, annot=True, cmap='mako')
```

from the heat map we can see that sleep duration and quality of sleep have a high positive correlation 

```python
sns.pairplot(sleep_analysis[['Sleep Duration', 'Quality of Sleep']],height=5, aspect = 1.5)
```

We can see from this that we have a decently spread out sleep duration
and most people had sleep quality of 6 and above

```python
#checking whats the rage of sleep 
sleep_analysis['Quality of Sleep'].unique()
```

```python
#group by sleep duration and get average sleep quality
plt.figure(figsize=(12,6))
avg_sleep_quality = sleep_analysis.groupby('Sleep Duration')['Quality of Sleep'].mean()
plt.plot(avg_sleep_quality.index, avg_sleep_quality.values, marker ='o', markersize = 4, linewidth = 1)

plt.xlabel('Sleep Duration (hours)')
plt.ylabel('Average Sleep Quality')
plt.title('Average Sleep Quality by Sleep Duration', fontsize = 18)

ax = plt.gca()
ax.spines["top"].set_visible(False)
ax.spines["right"].set_visible(False)

plt.tight_layout()
plt.show()
```

Not sure if there was a restriction on average sleep quality, but from this we can determine that  
sleep quality increases as sleep duration increases plateauing after 8 hours

```python
#checking unique values
sleep_analysis['Gender'].unique()
```

```python
#comparing how sleep duration affects sleep quality for male and female
df_gender = sleep_analysis[sleep_analysis['Gender'].isin(['Male','Female'])]
sns.pairplot(df_gender[['Sleep Duration','Quality of Sleep','Gender']], height =5, aspect = 1.5, diag_kind='kde', hue='Gender')
df_gender.shape[0]
```

from the bottom left we can see that females tend to have the same sleep quality with slightly less sleep duration than males
